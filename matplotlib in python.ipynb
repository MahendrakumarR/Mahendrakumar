```python
n=np.array([0,1,2,3,4,5])

fig, axes=plt.subplots(1,4, figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title("step")                         
axes[1].step(n, n**2, lw=2)

axes[2].set_title("bar")                                     # Draw third graph using bar  you can remove align,width,alpha then work.  alpha mean dark the bar    
axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5) 
```

```python
n=np.array([0,1,2,3,4,5])

fig, axes=plt.subplots(1,4, figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title("step")                         
axes[1].step(n, n**2, lw=2)

axes[2].set_title("bar")                                     
axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5) 

axes[3].set_title("fill_between")                                     # Draw fourth graph using fill between
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5) 
```

```python
n=np.random.randn(10000)
fig, axes=plt.subplots(1, 2, figsize=(12,4))

axes[0].set_title("Default Histogram")                           # The first graph have 8 bins draw
axes[0].hist(n)

axes[1].set_title("Cumulative Detailed  Histogram")
axes[1].hist(n, cumulative=True, bins=50)                        # The second graph have 50 bins 
```

```python
# Draw contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta=0.025
x=np.arange(-3.0, 3.0, delta)
y=np.arange(-2.0, 2.0, delta)
X,Y=np.meshgrid(x,y)
z1=np.exp(-X**2 - Y**2)
z2=np.exp(-(X-1)**2 - (Y-1)**2)
Z=(z1-z2)*2
print(X)
print(Y)
```

```python
fig,ax=plt.subplots()
cs=ax.contour(X,Y,Z)
ax.clabel(cs, inline=1, fontsize=10)
ax.set_title("Contour map")
```
